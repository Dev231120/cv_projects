Enriching the dataset with rainfall patterns,inventory suggestions,guided suggestions for stakeholders and risk analysis

```python
df_time_series['forecast_sarima'] = np.nan
df_time_series.loc[test.index, 'forecast_sarima'] = sarima_retail_forecast.values
```

```python
df_time_series['forecast_arima'] = np.nan
df_time_series.loc[test.index, 'forecast_arima'] = forecasted.values
```

```python
df_time_series.head()
```

```python
df_time_series['forecast_roll7'] = df_time_series['retail_price'].rolling(7).mean()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

def metrics(actual, pred):
    return {
        "MAE": mean_absolute_error(actual, pred),
        "RMSE": np.sqrt(mean_squared_error(actual, pred))
    }

metrics_sarima = metrics(test['retail_price'], df_time_series.loc[test.index, 'forecast_sarima'])
metrics_arima  = metrics(test['retail_price'], df_time_series.loc[test.index, 'forecast_arima'])
metrics_roll7  = metrics(test['retail_price'], df_time_series.loc[test.index, 'forecast_roll7'])
```

```python
metrics_sarima,metrics_arima,metrics_roll7
```

```python
df_time_series['final_forecast'] = df_time_series['forecast_roll7']
```

```python
df_time_series['inventory_action'] = np.where(
    df_time_series['final_forecast'] > df_time_series['retail_price'],
    'Hold Inventory',
    'Sell Inventory'
)
```

```python
df_time_series['rolling_30_avg'] = (
    df_time_series['retail_price']
    .rolling(30, min_periods=10)
    .mean()
)

df_time_series['procurement_signal'] = np.where(
    df_time_series['retail_price'] < df_time_series['rolling_30_avg'],
    'Buy Window',
    'Wait'
)
```

```python
df_time_series['volatility_30d'] = (
    df_time_series['retail_price']
    .rolling(30, min_periods=10)
    .std()
)
```

```python
low = df_time_series['volatility_30d'].quantile(0.33)
high = df_time_series['volatility_30d'].quantile(0.66)

df_time_series['risk_zone'] = np.where(
    df_time_series['volatility_30d'] <= low, 'Low Risk',
    np.where(
        df_time_series['volatility_30d'] <= high, 'Medium Risk',
        'High Risk'
    )
)
```

```python
df_time_series['final_recommendation'] = np.where(
    (df_time_series['inventory_action'] == 'Hold Inventory') &
    (df_time_series['risk_zone'] == 'Low Risk'),
    'Strong Hold',
    np.where(
        (df_time_series['inventory_action'] == 'Sell Inventory') &
        (df_time_series['risk_zone'] == 'High Risk'),
        'Urgent Sell',
        'Cautious'
    )
)
```
